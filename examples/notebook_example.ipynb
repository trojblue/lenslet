## Example 1: Basic Usage with Local Images

````python
import lenslet

# Example: Launch with a list of local image paths
datasets = {
    "my_photos": [
        "/path/to/photo1.jpg",
        "/path/to/photo2.png",
        "/path/to/photo3.jpg",
    ]
}

# Launch in non-blocking mode (returns immediately)
lenslet.launch(datasets, blocking=False, port=7070)

print("Gallery running at http://localhost:7070")
````

## Example 2: With Pandas DataFrame and S3


````python
import pandas as pd
import lenslet

# Example: DataFrame with mixed local and S3 images
df = pd.DataFrame({
    'image_path': [
        '/local/img1.jpg',
        's3://bucket/img2.jpg',
        '/local/img3.png',
    ],
    'label': ['cat', 'dog', 'cat']
})

# Launch with multiple datasets
datasets = {
    "all_images": df['image_path'].tolist(),
    "cats_only": df[df['label'] == 'cat']['image_path'].tolist(),
}

lenslet.launch(datasets, blocking=False, port=7071)
````
